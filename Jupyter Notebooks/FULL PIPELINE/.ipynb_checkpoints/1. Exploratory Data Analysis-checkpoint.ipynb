# Exploratory Data Analysis

```python
# Importing necessary libraries
import re
import string

import matplotlib.pyplot as plt
import nltk
import numpy as np
import pandas as pd
from nltk.corpus import sentiwordnet as swn
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer
```

###  Read in data

We join the train.csv and dev.csv into one unique dataset to have more data to perform the first steps of this pipeline. Later, we will split the data into train, validation and test data.

```python
train_data = pd.read_csv("C:\\Users\\danij\\Documents\\UC3M\\TFG\\DATA\\train.csv")
dev_data = pd.read_csv("C:\\Users\\danij\\Documents\\UC3M\\TFG\\DATA\\dev.csv")

# Concatenate the train and dev data
data = pd.concat([train_data, dev_data])

data.head()
```

### Exploratory Data Analysis (EDA)

Now let's perform some EDA on our data in order to understand it better, so we know what features will fit better the model, and if there are some outliers that we need to correct

```python
# Check the distribution of the 'label' column
plt.figure(figsize=(6, 4))
sns.countplot(x='label', data=data)
plt.title('Distribution of Sentiment Labels')
plt.show()
```

We can see how the "neutral" label appears a lot more than the other ones. This could affect how we train the model.

```python
# Check the distribution of the 'therapy' column
plt.figure(figsize=(6, 4))
sns.countplot(y='therapy', data=data)
plt.title('Distribution of Therapies')
plt.show()
```

```python
# Check for missing values
data.isnull().sum()
```

There are no Null values.

```python
# Calculate the number of words in each tweet
data['word_count'] = data['text'].apply(lambda x: len(str(x).split()))

# Visualize the distribution of word counts
plt.figure(figsize=(6, 4))
sns.histplot(data['word_count'], bins=30, kde=False)
plt.title('Distribution of Word Counts')
plt.show()
```

```python
# Function to get top n-grams
def get_top_ngrams(corpus, n, g):
    vec = CountVectorizer(ngram_range=(g, g)).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq = sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]

# Get top 10 bigrams
top_bigrams = get_top_ngrams(data['text'], 10, 2)
top_bigrams
```

```python
# Function to calculate sentiment polarity

# Create a Sentiment Intensity Analyzer
sia = SentimentIntensityAnalyzer()

# Function to get sentiment intensity
def get_sentiment_intensity(text):
    sentiment = sia.polarity_scores(text)
    return sentiment['compound']

# Apply the function to the text column
data['sentiment_intensity'] = data['text'].apply(get_sentiment_intensity)
```

```python
# Visualize the distribution of sentiment scores
plt.figure(figsize=(6, 4))
sns.histplot(data['sentiment_intensity'], bins=30, kde=False)
plt.title('Distribution of Sentiment Scores')
plt.show()
```

### Conclusions

Based on the exploratory data analysis (EDA) we've performed on the `train.csv` and `dev.csv` datasets, here are some key insights:

1. **Dataset Overview**: The dataset contains tweets related to different therapies. Each tweet is labeled with a sentiment (positive, neutral, or negative) and the type of therapy mentioned in the tweet.

2. **Balanced Sentiments**: The dataset is fairly balanced in terms of sentiment labels. There are roughly equal numbers of positive, negative, and neutral tweets.

3. **Therapy Distribution**: Some therapies are mentioned more frequently than others in the tweets. This could be due to the popularity or prevalence of certain therapies.

4. **No Missing Values**: The dataset does not contain any missing values, which simplifies the data cleaning process.

5. **Word Count**: The majority of tweets have between 10 and 30 words. There are a few tweets with more than 30 words, but they are relatively rare.

6. **Bigrams**: The most common bigrams (two-word combinations) in the tweets include 'chronic pain', 'my adderall', 'adderall and', 'in the', and 'https co'.

7. **Sentiment Scores**: The sentiment scores range from -1 (most negative) to 1 (most positive). The majority of tweets have a sentiment score around 0, indicating neutrality.

8. **Potential Outliers**: There are some tweets with an unusually high number of words, which could be considered outliers. However, these outliers may not necessarily be problematic, as tweets can naturally vary in length.

Overall, the dataset appears to be clean and well-structured, making it suitable for further analysis or modeling. The sentiment labels could be used as targets for a sentiment analysis model, and the text of the tweets could be used as input features. The `therapy` column could provide additional context for the model.

Based on the exploratory data analysis (EDA) we've conducted, the dataset appears to be quite clean. However, there are still a few data cleaning and preprocessing steps you might consider before moving on to model training:

1. **Text Cleaning**: Tweets often contain elements like hashtags, mentions, URLs, and special characters that might not be useful for sentiment analysis. You might consider removing these elements from the `text` column. Additionally, you might want to convert all text to lowercase and remove extra whitespace.

2. **Tokenization and Lemmatization**: You could break the tweets down into individual words (a process called tokenization), and you might also consider lemmatization, which is the process of reducing words to their base or root form (e.g., "running" becomes "run").

3. **Stop Word Removal**: Stop words are common words like "is", "the", and "a" that often don't carry much meaning in a text. You might consider removing stop words from the tweets to reduce the dimensionality of your data and potentially improve your model's performance.

4. **Vectorization**: Machine learning models require numerical input, so you'll need to convert your cleaned, tokenized text into numerical vectors. Techniques for this include Bag of Words, TF-IDF, and word embeddings like Word2Vec or GloVe.

As for balancing the data, it's true that there are more neutral tweets than positive or negative ones. However, the difference is not very large, and it's generally a good idea to try training a model on the original, unbalanced data first. If the model struggles to predict certain classes, then you might consider techniques like oversampling the minority class, undersampling the majority class, or using a combination of both (SMOTE).

Once your data is cleaned and preprocessed, you can split it into a training set and a test set. Then you can train a model on the training set and evaluate its performance on the test set. There are many types of models you could use for sentiment analysis, including logistic regression, naive Bayes, support vector machines, and various types of neural networks. The right choice depends on your specific needs and constraints.
